```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv('marketing.csv')
df.head()
```

```python
df = df.dropna()
df.info()
```

```python
sns.pairplot(df)
plt.show()
    
df.corr(numeric_only=True)
```

```python
df.corr(numeric_only=True)['Sales'].sort_values(ascending=False)
```

```python
X = df[['TV']]
X = sm.add_constant(X)
y = df['Sales']

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
residuals = model.resid

sns.histplot(residuals, kde=True)
plt.title("Residual Distribution")
plt.show()

plt.scatter(model.fittedvalues, residuals)
plt.axhline(0)
plt.title("Residuals vs Fitted")
plt.show()
```

```python
TV advertising has the strongest impact on sales.  
The model explains almost all variation (R² ≈ 0.999).  
Each increase in TV spend increases sales significantly.
```

```python
The company should invest more in TV advertising because it generates the highest return on investment.
```
